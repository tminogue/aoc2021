```python
import numpy as np
```

```python
# initialize input data

# bingo_file = open('./inputs/04_test_bingo.txt')
bingo_file = open('./inputs/04_bingo.txt')
bingo_raw = bingo_file.read().splitlines()

# parse number draw from boards text
draw_strings = bingo_raw[0].split(",")
draw = [int(i) for i in draw_strings]
# print(draw)
boards_raw = bingo_raw[2:]
boards_raw.append('') # add a blank string for parsing loop
# print(boards_raw[:10])

# parse each board from raw text
def get_board_arrays(boards_raw): 
    board_arrays = []
    board_array = []
    for idx, line in enumerate(boards_raw):
        if line != "":
            line_str_list = line.split()
            line_int_list = [int(i) for i in line_str_list]
            board_array.append(line_int_list)
        else:
            board_arrays.append(board_array)
            board_array = []
            continue
    return board_arrays

# parse each board into an 2-d array
def get_board_set(board_array):
    board_set = set()
    for row in board_array:
        for value in row:
            board_set.add(value)
    return board_set


# build dict of 
#   1. index
#   2. board array, 
#   3. set of values for checking presence of drawn number
#   4. result array for tracking position of drawn number

def build_board_dicts(board_arrays):
    board_dicts = []
    dim = len(board_arrays[0])
    for idx, board in enumerate(board_arrays):
        board_dicts.append({"id":idx, "board":board, 
                            "board set":get_board_set(board), 
                            "result":[[0 for x in range(dim)] for y in range(dim)],
                            "winner":False
                           })
    return(board_dicts)


# find winning patterns
def is_winner(board_result):
    
    dim=len(board_result)
    
    # check rows
    for row in board_result:
        if sum(row) == dim:
            # print(f"row winner")
            return True

    # check columms
    for i in range(dim):
        column=[]
        for row in board_result:
            column.append(row[i])
        if sum(column)==dim:
            # print(f"column winner")
            return True
    
#     #check diagonals
#     diagonal1 = [board_result[i][i] for i in range(dim)]
#     if sum(diagonal1)==dim:
#         print(f"diagonal winner left to right")
#         return True
    
#     diagonal2 = [board_result[i][dim-i-1] for i in range(dim)]
#     print(diagonal2)
#     if sum(diagonal2)==dim:
#         print(f"diagonal winner right to left")
#         return True
    
    else:
        return False
```

```python
def find_position(board, number):
    for idx, row in enumerate(board):
        if number in row:
            return (idx, row.index(number))

def find_first_winner(board_arrays):
    dim = 5
    all_boards = build_board_dicts(board_arrays)

    for idx, number in enumerate(draw):    
        for board_dict in all_boards:
            
            # check the set if the drawn number is present, only locate if present
            # and update result array with found position
            if number in board_dict["board set"]:
                position = find_position(board_dict["board"], number)
                board_dict["result"][position[0]][position[1]] = 1
                
            if idx >= dim-1:
#                 print(board_dict)
#                 print(is_winner(board_dict["result"]))
#                 print(number)
                if is_winner(board_dict["result"]):
                    board_number = board_dict["id"]
                    print(f"Board {board_number} is a winner in round {idx}")
                    return (board_dict["board"], board_dict["result"], number)
                
                
def find_last_winner(board_arrays):
    dim = 5
    all_boards = build_board_dicts(board_arrays)
    winners=[]

    for idx, number in enumerate(draw):
        
        for board_dict in all_boards:
            
            # check the set if the drawn number is present, only locate if present
            # and update result array with found position
            if number in board_dict["board set"]:
                position = find_position(board_dict["board"], number)
                board_dict["result"][position[0]][position[1]] = 1
                
            if idx >= dim-1:
#                 print(board_dict)
#                 print(is_winner(board_dict["result"]))
#                 print(number)
                if is_winner(board_dict["result"]) and board_dict["id"] not in (winners):
                    board_number = board_dict["id"]
                    print(f"Board {board_number} is winner in round {idx}")
                    board_dict["winner"]=True
                    if len(winners) < len(all_boards) - 1:
                        winners.append(board_dict["id"])
                    else:
                        return (board_dict["board"], board_dict["result"], number)
```

```python
def p1():
    board_arrays = get_board_arrays(boards_raw)
    winner = find_first_winner(board_arrays)
    # print(winner)

    final_number = winner[2]
    values_array = np.array(winner[0])
    marked_array = np.array(winner[1])
    # print(values_array)
    # print(marked_array)

    # invert 0 and 1 in marked array
    unmarked_array = (marked_array - 1) * -1
#     print(unmarked_array)

    product_array = np.multiply(values_array,unmarked_array)
#     print(product_array)

    product_values = []
#     print(product_values)

    for row in product_array:
        for value in row:
            product_values.append(value)
#     print(product_values)
#     print(sum(product_values))

    return sum(product_values) * final_number


def p2():
    board_arrays = get_board_arrays(boards_raw)
    winner = find_last_winner(board_arrays)
    # print(winner)

    final_number = winner[2]
    values_array = np.array(winner[0])
    marked_array = np.array(winner[1])
    # print(values_array)
    # print(marked_array)

    # invert 0 and 1 in marked array
    unmarked_array = (marked_array - 1) * -1
#     print(unmarked_array)

    product_array = np.multiply(values_array,unmarked_array)
#     print(product_array)

    product_values = []
#     print(product_values)

    for row in product_array:
        for value in row:
            product_values.append(value)
#     print(product_values)
#     print(sum(product_values))

    return sum(product_values) * final_number

```
